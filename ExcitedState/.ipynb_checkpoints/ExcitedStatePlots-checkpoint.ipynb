# PLOTS

## LEARNING CURVES

```python
def combined_LCS(data,krr_ref):
    markers = ['^','*','P','d']
    colors = pyl.cm.cividis(np.linspace(0,1,5))
    labels = ['SVP','6-31G','3-21G','STO-3G']
    lstyle = ['--','--','--','-','-','-']
    fig,ax = plt.subplots(2,3,figsize=(12,6),sharey=False)
    plt.tight_layout(h_pad=2)
    ax=ax.flatten('C')
    titles = ['Benzene','Naphthalene','Anthracene']
    n = 2**np.arange(1,10)
    ax[0].set_ylabel('MAE [meV]',fontsize=12)
    ax[3].set_ylabel('MAE [meV]',fontsize=12)
    y_minor = mticker.LogLocator(base = 10, subs = np.arange(1.0,20,3), numticks = 20)
    y_major = mticker.LogLocator(base = 10, subs = np.arange(1.0,10,3), numticks = 10)

    for i in range(6):
        ax[i].loglog(n,1e3*krr_ref[i],marker='o',color=colors[0])
        for j in range(4):
            ax[i].loglog(n,1e3*data[i][3-j,:],marker=markers[3-j],color=colors[j+1], 
                         linestyle=lstyle[i])
        if i<3:
            ax[i].set_title(titles[i],fontsize=15)
        ax[i].xaxis.set_major_formatter(mticker.ScalarFormatter())
        ax[i].set_xticks([2,8,32,128,512])
        ax[i].yaxis.set_minor_formatter(mticker.ScalarFormatter())
        ax[i].yaxis.set_major_formatter(mticker.ScalarFormatter())
        ax[i].yaxis.set_minor_locator(y_minor)
        ax[i].yaxis.set_major_locator(y_major)
        ax[i].grid(True, which="major", ls="-",color='dimgray')
        ax[i].grid(True, axis='y', which="minor", ls="--",color='lightgray')
        if i>2:
            ax[i].set_xlabel('$N_{\mathrm{train}}^{\mathrm{TZVP}}$',fontsize=12)
        ax[i].yaxis.set_tick_params(labelleft=True,which='both')
    fig.legend(['KRR-reference','SVP','6-31G','3-21G','STO-3G'],ncol=5,
               bbox_to_anchor=(0.75,-0.02),title='$f_b$')
    #DFTB
    #ax[3].sharey(ax[0])
    #ax[4].sharey(ax[1])
    #ax[5].sharey(ax[2])
    #CMD 
    ax[0].sharey(ax[3])
    ax[1].sharey(ax[4])
    ax[2].sharey(ax[5])
```

```python
combined_LCS([bencmd[0],napcmd[0],antcmd[0],bencmd[1],napcmd[1],antcmd[1]],
             [sf_bencmd,sf_napcmd,sf_antcmd,sf_bencmd,sf_napcmd,sf_antcmd])
```

## COEFF STUDY

```python
def coeff_subspace(OLScoef, ax = None):
    nfids=5
    ax = ax or plt.gca()
    m=40
    plt.tight_layout()
    labels = np.asarray(['STO-3G','3-21G',
                         '6-31G','SVP','TZVP'])
    
    colors = pyl.cm.cividis(np.linspace(0,1,9))
    
    
    count_arr = np.asarray([0,0,1,1,2,2,3,3,4])
    ind_arr = np.asarray([13,12,12,11,11,10,10,9,9])
    
    count1 = 0
    for i in range(nfids-1,-1,-1):
        
        ax.plot([i+1+8],[labels[count1]],marker='s',color=colors[nfids-1+i],markersize=40,linestyle='')
        if i>0:
            ax.plot([i+8],[labels[count1]],marker='s',color=colors[i],markersize=40,linestyle='')
            
        count1 += 1
    ax.set_ylim(-0.5,4.5)
    ax.set_xlim(8.5,13.5)
    ax.set_xlabel('$\log_2(N_{\mathrm{train}})$',fontsize=12)
    ax.tick_params(axis='both', which='major', labelsize=12)
    ax.grid(True, which="major", ls="--",color='silver')
    
    
    #placing the coeffs as text
    col_arr = ['k','k','w','w','w','k','k','w','w']
    for i in range(5):
        ax.text(x=12.75-i,y=labels[i],s=str(np.round(OLScoef[-1][i],2)),fontsize=12,color='k',weight='bold')
        try:
            ax.text(x=11.75-i,y=labels[i],s=str(np.round(OLScoef[-1][5+i],2)),fontsize=12,color='w',weight='bold')
        except:
            pass
```

```python
fig,ax = plt.subplots(1,3,figsize=(15,5))
plt.tight_layout(w_pad=4)
coeff_subspace(coo_bencmd[0],ax[0])
coeff_subspace(coo_napcmd[0],ax[1])
coeff_subspace(coo_antcmd[0],ax[2])

ax[0].set_title('Benzene',fontsize=15)
ax[1].set_title('Naphthalene',fontsize=15)
ax[2].set_title('Anthracene',fontsize=15)
```

```python
fig,ax = plt.subplots(1,3,figsize=(15,5))
plt.tight_layout(w_pad=4)
coeff_subspace(coo_bendftb[0],ax[0])
coeff_subspace(coo_napdftb[0],ax[1])
coeff_subspace(coo_antdftb[0],ax[2])

ax[0].set_title('Benzene',fontsize=15)
ax[1].set_title('Naphthalene',fontsize=15)
ax[2].set_title('Anthracene',fontsize=15)
```
